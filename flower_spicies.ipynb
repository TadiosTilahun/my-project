```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbo
pd.set_option("display.max_rows", None)
```

```python
df = pd.read_csv("/Users/tadiostilahun/visoal studio/Iris.csv", index_col= "Id")
df
```

in the last columns change describtion:
    "Iris-setosa" reaplase to int number 0.
    "Iris-versicolor" reaplase to int number 1.
    "Iris-virginica" reaplase to int number 2.

```python
df[df.columns[-1]].replace(["Iris-setosa", "Iris-versicolor", "Iris-virginica"], [0,1,2], inplace= True)
```

```python
df.corr()
```

```python
df0 = df.loc[df["Species"] == 0]
df1 = df.loc[df["Species"] == 1]
df2 = df.loc[df["Species"] == 2]
for i in df.columns[1:-1]:
    plt.scatter(df0[df0.columns[0]], df0[i])
    plt.scatter(df1[df1.columns[0]], df1[i])
    plt.scatter(df2[df2.columns[0]], df2[i])
    plt.xlabel(df.columns[0])
    plt.ylabel(i)
    plt.legend()
    plt.show()
for j in df.columns[2:-1]:
    plt.scatter(df0[df0.columns[1]], df0[j])
    plt.scatter(df1[df1.columns[1]], df1[j])
    plt.scatter(df2[df2.columns[1]], df2[j])
    plt.xlabel(df.columns[1])
    plt.ylabel(j)
    plt.legend()
    plt.show()
plt.scatter(df0[df0.columns[2]], df0[df0.columns[3]])
plt.scatter(df1[df1.columns[2]], df1[df1.columns[3]])
plt.scatter(df2[df2.columns[2]], df2[df2.columns[3]])
plt.xlabel(df.columns[2])
plt.ylabel(df.columns[3])
plt.legend()
plt.show()
```

```python
x = df[df.columns[:-1]]
y = df[df.columns[-1]]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size= 0.66, random_state= 33)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
```

at this point i don't know what scoring algmorith to use, or choosing a scoring algmorith is not nussusery what ever but it lookes like my model is very good 

```python
y_predict = knn.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test, y_predict)
```
