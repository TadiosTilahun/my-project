```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sbo
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("/Users/tadiostilahun/visoal studio/Salary.csv")
df
```

```python
plt.scatter(df["YearsExperience"], df["Salary"])
plt.show()
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,1]
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, train_size= 0.66, random_state= 10)
train_y
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_x, train_y)
```

```python
pridict_y = lr.predict(test_x)
from sklearn.metrics import r2_score
r2_score(test_y, pridict_y)
```
