```python
df.drop(["fractal_dimension_mean", "texture_se", "smoothness_se", "symmetry_se", "Unnamed: 32"], axis= 1, inplace= True)
df.columns
```

```python
from sklearn.model_selection import train_test_split
x = df.iloc[:,2:]
y = df.iloc[:,1]
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size= 0.7, random_state= 55)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

not a bad model.

```python
y_pridict = lr.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test, y_pridict)
```
